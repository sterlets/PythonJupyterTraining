```python
def myFkt2(x):
    return x**2

def myFkt3(x):
    return x**3
```

```python
def floatRange(start, stop, step):
    output = [start]
    nOfPoints = (stop-start)/step
    while start <= stop:
        start = start + step
        output.append(start)
    return output
```

```python
xValues = floatRange(-1.5,1.5,0.1)
```

```python
def mapFkt(fkt, array):
    result = []
    for x in array:
        result.append(fkt(x))
    return result

```

```python
yValues2 = mapFkt(myFkt2, xValues)
yValues3 = mapFkt(myFkt3, xValues)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(xValues, yValues2)
plt.plot(xValues, yValues3)
plt.xlabel("x Werte")
plt.xlabel("y Werte")
plt.title("Plot Name")
```
